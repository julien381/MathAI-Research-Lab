```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)


df = pd.read_csv("../data/raw/creditcard.csv")
print("Fichier chargé avec succès !")
```

```python
# Vérification des valeurs manquantes (on cherche à voir des 0 partout)
df.isnull().sum()
```

```python
# Récupération de la colonne 'Class' et comptage des transactions normales (0) et des fraudes (1)
fraudes = df["Class"].value_counts()
print(fraudes)
```

```python
# Calculer le pourcentage exact de fraude
fraud_rate = df["Class"].mean() * 100
print(f"Taux de fraude : {fraud_rate:.4f}%")
```

```python
# Graphique en barres pour voir le déséquilibre géant
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="Class")
plt.title("Distribution des transactions (0 = Normal, 1 = Fraude)")
plt.xlabel("Classe")
plt.ylabel("Nombre de transactions")
plt.show()
```

```python
# Matrice de corrélation pour voir quelles variables sont liées
corr = df.corr()

plt.figure(figsize=(14,10))
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Matrice de corrélation globale")
plt.show()
```

```python
# Distribution des montants des transactions
plt.figure(figsize=(8,4))
sns.histplot(df["Amount"], bins=50)
plt.title("Distribution des montants des transactions")
plt.show()
```
